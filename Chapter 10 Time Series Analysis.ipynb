# Finding the last time crime was 20% lower with merge_asof

```python
muqadas_crime_sort = pd.read_hdf('C:\\Users\Documents\\crime.h5', 'crime') \
               .set_index('REPORTED_DATE') \
               .sort_index()
```

```python
all_data = muqadas_crime_sort.groupby([pd.Grouper(freq='M'), 'OFFENSE_CATEGORY_ID']).size()
all_data.head()
```

```python
all_data = all_data.sort_values().reset_index(name='Total')
all_data.head()
```

```python
goal = all_data[all_data['REPORTED_DATE'] == '2017-8-31'].reset_index(drop=True)
goal['Total_Goal'] = goal['Total'].mul(.8).astype(int)
goal.head()
```

```python
goal['Total_Goal'] = goal['Total_Goal'].astype('int64')
all_data['Total'] = all_data['Total'].astype('int64')
pd.merge_asof(
    goal, 
    all_data, 
    left_on='Total_Goal', 
    right_on='Total', 
    by='OFFENSE_CATEGORY_ID', 
    suffixes=('_Current', '_Last')
)
```
